```python
import numpy as np

# Original activities list
activities = [
    ('wkday_homework_hr', 'wkday_homework_min'),
    ('wkend_homework_hr', 'wkend_homework_min'),
    ('wkday_assign_hr', 'wkday_assign_min'),
    ('wkday_cram_hr', 'wkday_cram_min'),
    ('wkday_sports_hr', 'wkday_sports_min'),
    ('wkday_reading_hr', 'wkday_reading_min'),
    ('wkday_tv_hr', 'wkday_tv_min'),
    ('wkday_gaming_hr', 'wkday_gaming_min'),
    ('wkend_assign_hr', 'wkend_assign_min'),
    ('wkend_cram_hr', 'wkend_cram_min'),
    ('wkend_sports_hr', 'wkend_sports_min'),
    ('wkend_reading_hr', 'wkend_reading_min'),
    ('wkend_tv_hr', 'wkend_tv_min'),
    ('wkend_gaming_hr', 'wkend_gaming_min'),
    ('avg_sleep_hr_night', 'avg_sleep_min_night')
]

# Calculate total minutes for each activity and add weekdays and weekends together
for hr_col, min_col in activities:
    total_min_col = hr_col.replace('_hr', '_total_min')
    
    # Replace missing values in either hours or minutes with 0
    df[total_min_col] = df.apply(
        lambda row: (row[hr_col] if not pd.isna(row[hr_col]) else 0) * 60 +
                    (row[min_col] if not pd.isna(row[min_col]) else 0)
        if not (pd.isna(row[hr_col]) and pd.isna(row[min_col]))
        else np.nan,
        axis=1
    )

# Add combined weekly activity time
for activity in set(col.split('_')[1] for col, _ in activities if 'wkday' in col):
    wkday_col = f'wkday_{activity}_total_min'
    wkend_col = f'wkend_{activity}_total_min'
    combined_col = f'weekly_{activity}_total_min'

    if wkday_col in df.columns and wkend_col in df.columns:
        df[combined_col] = (5 * df[wkday_col] + 2 * df[wkend_col]) / 7

# Add combined categories for study and recreation
study_activities = ['homework', 'assign', 'cram']
recreation_activities = ['sports', 'reading', 'tv', 'gaming']

# Weekly study time
df['weekly_study_total_min'] = df[[f'weekly_{activity}_total_min' for activity in study_activities]].sum(axis=1)

# Weekly recreation time
df['weekly_recreation_total_min'] = df[[f'weekly_{activity}_total_min' for activity in recreation_activities]].sum(axis=1)

# Drop original columns
columns_to_drop = [
    'wkday_homework_hr', 'wkday_homework_min',
    'wkend_homework_hr', 'wkend_homework_min',
    'wkday_assign_hr', 'wkday_assign_min',
    'wkday_cram_hr', 'wkday_cram_min',
    'wkday_sports_hr', 'wkday_sports_min',
    'wkday_reading_hr', 'wkday_reading_min',
    'wkday_tv_hr', 'wkday_tv_min',
    'wkday_gaming_hr', 'wkday_gaming_min',
    'wkend_assign_hr', 'wkend_assign_min',
    'wkend_cram_hr', 'wkend_cram_min',
    'wkend_sports_hr', 'wkend_sports_min',
    'wkend_reading_hr', 'wkend_reading_min',
    'wkend_tv_hr', 'wkend_tv_min',
    'wkend_gaming_hr', 'wkend_gaming_min',
    'avg_sleep_hr_night', 'avg_sleep_min_night',
    'wkday_homework_total_min', 'wkend_homework_total_min',
    'wkday_assign_total_min', 'wkend_assign_total_min',
    'wkday_cram_total_min', 'wkend_cram_total_min',
    'wkday_sports_total_min', 'wkend_sports_total_min',
    'wkday_reading_total_min', 'wkend_reading_total_min',
    'wkday_tv_total_min', 'wkend_tv_total_min',
    'wkday_gaming_total_min', 'wkend_gaming_total_min',
]

df.drop(columns=columns_to_drop, inplace=True)

# Print to verify
print(df[['weekly_study_total_min', 'weekly_recreation_total_min']].head())


```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer

onehot_ftrs = [
    'grade9', 'gender', 'ethnicity', 'birthplace', 'hukou_loc', 'hukou_type',
    'current_residence',  'short_sighted', 'hospitalized_last_yr',
     'only_child', 'sibling_school_status', 'live_with_mother',
    'live_with_father', 'live_with_siblings', 'live_with_grandparents',
    'live_with_relatives', 'live_with_nonrelatives', 'not_live_with_mother',
    'not_live_with_father', 'not_live_with_siblings', 'family_live_together',
    'mother_occupation', 'father_occupation', 'father_drinks', 'parents_quarrel',
    'parents_get_along', 'own_desk', 'no_extra_activities', 'math_olympiad',
    'ord_math', 'chinese_composition', 'study_english', 'paint_draw', 'calligraphy',
    'music_instrument', 'dancing', 'play_chess', 'sports', 'other_courses',
    'no_hobbies', 'play_instruments', 'vocal_dance_act', 'hobby_calligraphy',
    'hobby_paint_draw', 'play_chess_hobby', 'sports_hobby', 'other_hobbies',
    'cram_courses', 'attend_camps', 'help_parents', 'visit_hometown', 'stay_home',
    'other_vacation_activities', 'chat_when_needed',
    'trouble_help', 'need_help', 'language_with_parents','illness_pre_school','computer_access'
]

ordinal_ftrs = [
    'birthweight_cat','current_health','homework_checkup', 'homework_guidance',
    'concern_homework', 'concern_behavior', 'concern_attendance', 'concern_home_arrival_time',
    'concern_friends', 'concern_dress_style', 'concern_internet_time', 'concern_tv_time',
    'school_events_mother', 'school_events_father', 'friends_relation_mother',
    'friends_relation_father', 'teacher_relation_mother', 'teacher_relation_father',
    'feelings_mother', 'feelings_father', 'troubles_mother', 'troubles_father',
    'relation_with_mother', 'relation_with_father', 'fin_cond_pre_school', 'mother_education', 'father_education',
    'current_fin_cond', 'books_owned', 'visit_museums_freq', 'watch_shows_freq',
    'family_dinner_freq', 'family_reading_time', 'family_tv_time', 'family_sports_time',
    'visit_science_museums', 'watch_movies_freq', 'parents_contribution_perception',
    'parents_academic_expectation', 'parents_expected_education_level', 'feel_about_expectations',
    'expected_future_job', 'expected_living_area', 'parents_confidence'
]
std_ftrs = [
    'stcog', 'stdchn', 'stdmat', 'stdeng', 'birth_year', 'height_cm', 'weight_kg',
    'avg_sleep_total_min_night',
    'weekly_study_total_min',
    'weekly_recreation_total_min',
]



preprocessor = ColumnTransformer(
    transformers=[
        ('ord', Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value=0)),  # Impute missing values with 0
            ('ordinal', OrdinalEncoder())  # Ordinal encoding
        ]), ordinal_ftrs),
        ('onehot', Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value=-1)),  # Impute missing values with 'missing'
            ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))  # One-hot encoding
        ]), onehot_ftrs),
        ('std', StandardScaler(), std_ftrs)])
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split, KFold, GridSearchCV,ParameterGrid
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import pandas as pd
from sklearn.base import clone
def MLpipe_ReducedFeature_R2(X, y, preprocessor, ML_algo, param_grid):
    """
    Perform hyperparameter tuning using random states, dynamically handling missing value patterns.
    Use the original submodel pattern logic, with R as the evaluation metric.
    """
    r2_scores = []
    final_models = []
    for i in [1,2,3,4,5]:
        print(f"Random State {i}")

        # let's split to train, CV, and test
        X_other, X_test, y_other, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
        X_train, X_CV, y_train, y_CV = train_test_split(X_other, y_other, test_size=0.25, random_state=i)

        print(X_train.shape)
        print(X_CV.shape)
        print(X_test.shape)

    # fit_transform the training set
        X_prep = preprocessor.fit_transform(X_train)
# collect feature names
        feature_names = preprocessor.get_feature_names_out()

        df_train = pd.DataFrame(data=X_prep,columns=feature_names)
        print(df_train.shape)

# transform the CV
        df_CV = preprocessor.transform(X_CV)
        df_CV = pd.DataFrame(data=df_CV,columns = feature_names)
        print(df_CV.shape)

# transform the test
        df_test = preprocessor.transform(X_test)
        df_test = pd.DataFrame(data=df_test,columns = feature_names)
        print(df_test.shape)

        X_train=df_train
        Y_train=y_train
        X_CV=df_CV
        X_test=df_test

        # find all unique patterns of missing value in test set
        mask = X_test.isnull()
        unique_rows = np.array(np.unique(mask, axis=0))
        all_y_test_pred = pd.DataFrame()
    
        print('there are', len(unique_rows), 'unique missing value patterns.')
    
    # divide test sets into subgroups according to the unique patterns
        for i in range(len(unique_rows)):
            print ('working on unique pattern', i)
        ## generate X_test subset that matches the unique pattern i
            sub_X_test = pd.DataFrame()
            sub_y_test = pd.Series(dtype=float)
            for j in range(len(mask)): # check each row in mask
                row_mask = np.array(mask.iloc[j])
                if np.array_equal(row_mask, unique_rows[i]): # if the pattern matches the ith unique pattern

                    sub_X_test = pd.concat([sub_X_test,X_test.iloc[[j]]])# append the according X_test row j to the subset
                    sub_y_test = pd.concat([sub_y_test, y_test.iloc[[j]]])# append the according y_test row j

            sub_X_test = sub_X_test[X_test.columns[~unique_rows[i]]]
        
        ## choose the according reduced features for subgroups
            sub_X_train = pd.DataFrame()
            sub_Y_train = pd.DataFrame()
            sub_X_CV = pd.DataFrame()
            sub_y_CV = pd.DataFrame()
        # 1.cut the feature columns that have nans in the according sub_X_test
            sub_X_train = X_train[X_train.columns[~unique_rows[i]]]
            sub_X_CV = X_CV[X_CV.columns[~unique_rows[i]]]
        # 2.cut the rows in the sub_X_train and sub_X_CV that have any nans
            sub_X_train = sub_X_train.dropna()
            sub_X_CV = sub_X_CV.dropna()   
        # 3.cut the sub_Y_train and sub_y_CV accordingly
            sub_Y_train = Y_train.iloc[sub_X_train.index]
            sub_y_CV = y_CV.iloc[sub_X_CV.index]

            # Hyperparameter tuning for the current pattern
            val_scores = []
            models = []

            for params in ParameterGrid(param_grid):
                print(f"      Hyperparameters: {params}")
                model = clone(ML_algo).set_params(**params)
                model.fit(sub_X_train, sub_Y_train)
                models.append(model)
                # Evaluate on validation set
                val_r2 = r2_score(sub_y_CV, model.predict(sub_X_CV))
                val_scores.append(val_r2)
                print(f"         Validation R: {val_r2:.4f}")

            # Select the best model for the current pattern
            best_idx = np.argmax(val_scores)
            best_model = models[best_idx]
            print(f"      Best Parameters for Pattern {i}: {ParameterGrid(param_grid)[best_idx]}")
            print(f"      Best Validation R: {val_scores[best_idx]:.4f}")
            # Evaluate on the test set for the current pattern
            sub_y_test_pred = pd.Series(best_model.predict(sub_X_test), index=sub_y_test.index)

            # Convert to DataFrame
            sub_y_test_pred = pd.DataFrame(sub_y_test_pred, columns=['sub_y_test_pred'])
            all_y_test_pred = pd.concat([all_y_test_pred, sub_y_test_pred])


        # Sort predictions and calculate overall R for the current random state
        all_y_test_pred = all_y_test_pred.sort_index()
        y_test = y_test.sort_index()

        total_r2 = r2_score(y_test, all_y_test_pred)
        print(f"   Final R: {total_r2:.4f}")

        r2_scores.append(total_r2)
        final_models.append(best_model)

    # Summarize results
    mean_r2 = np.mean(r2_scores)
    std_r2 = np.std(r2_scores)

    print(f"\nFinal Results Across All Random States:")
    print(f"Mean R: {mean_r2:.4f}  {std_r2:.4f}")

    return r2_scores, final_models
```

```python
import xgboost
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import r2_score
r2_scores = []
final_models = []
y = df['y']
X = df.loc[:, df.columns != 'y']
# Drop observations with missing y
non_missing_mask = ~y.isnull()
X = X[non_missing_mask]
y = y[non_missing_mask]
for i in [1,2,3,4,5]:
        print(f"Random State {i}")

        # let's split to train, CV, and test
        X_other, X_test, y_other, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
        X_train, X_CV, y_train, y_CV = train_test_split(X_other, y_other, test_size=0.25, random_state=i)

        print(X_train.shape)
        print(X_CV.shape)
        print(X_test.shape)

    # fit_transform the training set
        X_prep = preprocessor.fit_transform(X_train)
# collect feature names
        feature_names = preprocessor.get_feature_names_out()

        df_train = pd.DataFrame(data=X_prep,columns=feature_names)
        print(df_train.shape)

# transform the CV
        df_CV = preprocessor.transform(X_CV)
        df_CV = pd.DataFrame(data=df_CV,columns = feature_names)
        print(df_CV.shape)

# transform the test
        df_test = preprocessor.transform(X_test)
        df_test = pd.DataFrame(data=df_test,columns = feature_names)
        print(df_test.shape)

        XGB = xgboost.XGBRegressor(random_state=i,n_jobs=1)
        param_grid = {"learning_rate": [0.03],
                      "n_estimators": [10000],
                      "seed": [0],
                      "reg_alpha": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
                      "reg_lambda": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
                      "missing": [np.nan], 
                      "max_depth": [1,3,10,30,100],
                      "colsample_bytree": [0.9],              
                      "subsample": [0.66]}
        pg = ParameterGrid(param_grid)
        scores = np.zeros(len(pg))
        for i in range(len(pg)):
            params = pg[i]
            XGB.set_params(**params,early_stopping_rounds=50)
            eval_set = [(df_CV, y_CV)]
            XGB.fit(df_train, y_train, eval_set=eval_set, verbose=False)# with early stopping
            y_CV_pred = XGB.predict(df_CV)
            scores[i] = r2_score(y_CV,y_CV_pred)
            print(params)
            print('the CV R2:',r2_score(y_CV,y_CV_pred))
        best_params = np.array(pg)[scores == np.max(scores)]
        print('Validation set max score and best parameters are:')
        print(np.max(scores))
        print(best_params)
        XGB.set_params(**best_params[0],early_stopping_rounds=50)
        XGB.fit(df_train,y_train,eval_set=eval_set, verbose=False)
        y_test_pred = XGB.predict(df_test)

        test_r2 = r2_score(y_test, y_test_pred)
        r2_scores.append(test_r2)
        final_models.append(XGB)

mean_r2 = np.mean(r2_scores)
std_r2 = np.std(r2_scores)

print(f"\nFinal Results Across All Random States:")
print(f"Mean R: {mean_r2:.4f}  {std_r2:.4f}")
```
