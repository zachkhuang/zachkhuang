### Data Preprocessing and Splitting

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer

onehot_ftrs = [
    'grade9', 'gender', 'ethnicity', 'birthplace', 'hukou_loc', 'hukou_type',
    'current_residence',  'short_sighted', 'hospitalized_last_yr',
     'only_child', 'sibling_school_status', 'live_with_mother',
    'live_with_father', 'live_with_siblings', 'live_with_grandparents',
    'live_with_relatives', 'live_with_nonrelatives', 'not_live_with_mother',
    'not_live_with_father', 'not_live_with_siblings', 'family_live_together',
    'mother_occupation', 'father_occupation', 'father_drinks', 'parents_quarrel',
    'parents_get_along', 'own_desk', 'no_extra_activities', 'math_olympiad',
    'ord_math', 'chinese_composition', 'study_english', 'paint_draw', 'calligraphy',
    'music_instrument', 'dancing', 'play_chess', 'sports', 'other_courses',
    'no_hobbies', 'play_instruments', 'vocal_dance_act', 'hobby_calligraphy',
    'hobby_paint_draw', 'play_chess_hobby', 'sports_hobby', 'other_hobbies',
    'cram_courses', 'attend_camps', 'help_parents', 'visit_hometown', 'stay_home',
    'other_vacation_activities', 'chat_when_needed',
    'trouble_help', 'need_help', 'language_with_parents','illness_pre_school',
]

ordinal_ftrs = [
    'birthweight_cat','current_health','homework_checkup', 'homework_guidance',
    'concern_homework', 'concern_behavior', 'concern_attendance', 'concern_home_arrival_time',
    'concern_friends', 'concern_dress_style', 'concern_internet_time', 'concern_tv_time',
    'school_events_mother', 'school_events_father', 'friends_relation_mother',
    'friends_relation_father', 'teacher_relation_mother', 'teacher_relation_father',
    'feelings_mother', 'feelings_father', 'troubles_mother', 'troubles_father',
    'relation_with_mother', 'relation_with_father', 'fin_cond_pre_school', 'mother_education', 'father_education',
    'current_fin_cond', 'books_owned', 'visit_museums_freq', 'watch_shows_freq',
    'family_dinner_freq', 'family_reading_time', 'family_tv_time', 'family_sports_time',
    'visit_science_museums', 'watch_movies_freq', 'parents_contribution_perception',
    'parents_academic_expectation', 'parents_expected_education_level', 'feel_about_expectations',
    'expected_future_job', 'expected_living_area', 'parents_confidence'
]
std_ftrs = [
    'stcog', 'stdchn', 'stdmat', 'stdeng', 'birth_year', 'age_arrival', 'height_cm', 'weight_kg',
    'avg_sleep_total_min_night', 'wkday_homework_total_min', 'wkend_homework_total_min',
    'wkday_assign_total_min', 'wkday_cram_total_min', 'wkday_sports_total_min', 'wkday_reading_total_min',
    'wkday_tv_total_min', 'wkday_gaming_total_min', 'wkday_housework_total_min', 'wkend_assign_total_min',
    'wkend_cram_total_min', 'wkend_sports_total_min', 'wkend_reading_total_min', 'wkend_tv_total_min',
    'wkend_gaming_total_min', 'wkend_housework_total_min'
]

preprocessor = ColumnTransformer(
    transformers=[
        ('ord', Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value=0)),  # Impute missing values with 0
            ('ordinal', OrdinalEncoder())  # Ordinal encoding
        ]), ordinal_ftrs),
        ('onehot', Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value=-1)),  # Impute missing values with 'missing'
            ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))  # One-hot encoding
        ]), onehot_ftrs),
        ('std', StandardScaler(), std_ftrs)])

clf = Pipeline(steps=[('preprocessor', preprocessor)])
```

```python
y=df['y']
X = df.loc[:, df.columns != 'y']
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import KFold
random_state =42

# first split to separate out the test set
X_other, X_test, y_other, y_test = train_test_split(X,y,test_size = 0.2,random_state=random_state)
print(X_other.shape,y_other.shape)
print('test set:',X_test.shape,y_test.shape)

# do KFold split on other
kf = KFold(n_splits=4,shuffle=True,random_state=random_state)
for train_index, val_index in kf.split(X_other,y_other):
    X_train = X_other.iloc[train_index]
    y_train = y_other.iloc[train_index]
    X_val = X_other.iloc[val_index]
    y_val = y_other.iloc[val_index]
    print('   training set:',X_train.shape, y_train.shape) 
    print('   validation set:',X_val.shape, y_val.shape) 
    X_train_prep = clf.fit_transform(X_train)
    X_val_prep = clf.transform(X_val)
    X_test_prep = clf.transform(X_test)
    print('Training set:', X_train_prep.shape, y_train.shape) 
    print('Validation set:', X_val_prep.shape, y_val.shape)
    feature_names = preprocessor.get_feature_names_out()
    df_Xtrain = pd.DataFrame(X_train_prep, columns=feature_names)
    print(df_Xtrain.columns)
    print('data dimensions:',df_Xtrain.shape)
    perc_missing_per_ftr = df_Xtrain.isnull().sum(axis=0)/df_Xtrain.shape[0]
    print('fraction of missing values in features:')
    print(perc_missing_per_ftr[perc_missing_per_ftr > 0],len(perc_missing_per_ftr[perc_missing_per_ftr > 0]))
    print('data types of the features with missing values:')
    print(df_Xtrain[perc_missing_per_ftr[perc_missing_per_ftr > 0].index].dtypes)
    frac_missing = sum(df_Xtrain.isnull().sum(axis=1)!=0)/df_Xtrain.shape[0]
    print('fraction of points with missing values:',frac_missing)
```
